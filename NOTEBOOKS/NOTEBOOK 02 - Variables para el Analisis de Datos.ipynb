## 1. Variables con significado

En un algoritmo de análisis de datos podemos describir cada variable desde
**dos perspectivas diferentes**:

### A. Según su origen

| Categoría | Función | Ejemplo |
|---|---|---|
| Dato original | Conserva información recibida | `ventas_diarias` |
| Constante de referencia | Representa un valor que no debe cambiar durante el análisis | `META_SEMANAL` |
| Variable derivada | Se calcula utilizando otros datos | `total_ventas` |

### B. Según su propósito

| Categoría | Función | Ejemplo |
|---|---|---|
| Indicador | Resume o interpreta los datos | `promedio_diario` |
| Variable lógica | Contesta una pregunta con `True` o `False` | `meta_cumplida` |
| Salida interpretada | Comunica una conclusión | `conclusion` |

Una variable puede pertenecer a una categoría de cada perspectiva. Por ejemplo,
`promedio_diario` es una **variable derivada por su origen** y un
**indicador por su propósito**.

Usaremos mayúsculas para constantes y `snake_case` para las demás variables.

````python
# Ejemplo guiado: variables de un análisis comercial
ventas_diarias = [125000, 98000, 147000, 132000, 110000]
META_SEMANAL = 600000

total_ventas = sum(ventas_diarias)
cantidad_dias = len(ventas_diarias)
promedio_diario = total_ventas / cantidad_dias
meta_cumplida = total_ventas >= META_SEMANAL

print(f"Total de ventas: ₡{total_ventas:,.2f}")
print(f"Promedio diario: ₡{promedio_diario:,.2f}")
print(f"¿Se cumplió la meta?: {meta_cumplida}")
````

## 2. Dependencias y trazabilidad

Una variable derivada depende de otras variables. El orden del algoritmo es
importante porque Python necesita conocer los datos antes de calcular un
resultado.

```text
ventas_diarias ----> total_ventas ----+
       |                              |
       +----> cantidad_dias ----------+----> promedio_diario
       |
       +----> total_ventas ----+
                               +----> meta_cumplida
META_SEMANAL ------------------+
```

El promedio depende de `total_ventas` y `cantidad_dias`. La decisión depende de
`total_ventas` y `META_SEMANAL`.

Si cambia una venta, deben recalcularse todos los indicadores que dependen de
ella.

## 3. De datos a indicadores

Una lista permite reunir valores relacionados:

```python
ventas_diarias = [125000, 98000, 147000, 132000, 110000]
```

Sin utilizar ciclos podemos obtener indicadores:

| Función | Resultado |
|---|---|
| `sum(datos)` | Suma de los valores |
| `len(datos)` | Cantidad de valores |
| `max(datos)` | Valor mayor |
| `min(datos)` | Valor menor |

El **rango** es la diferencia entre el valor mayor y el menor. Un rango elevado
indica mayor variación entre los datos.

### Actividad 3.1 - Construcción guiada del reporte

Complete los indicadores. Antes de programar, clasifique cada variable como
dato original, variable derivada o indicador.

````python
ventas_diarias = [125000, 98000, 147000, 132000, 110000]

# Complete las fórmulas reemplazando los puntos suspensivos.
# La celda producirá un error hasta que todas las fórmulas estén correctas.
total_ventas = sum(ventas_diarias)
cantidad_dias = len(ventas_diarias)
promedio_diario = total_ventas / cantidad_dias
venta_mayor = max(ventas_diarias)
venta_menor = min(ventas_diarias)
rango_ventas = venta_mayor - venta_menor


assert (
    total_ventas is not Ellipsis
    and cantidad_dias is not Ellipsis
    and promedio_diario is not Ellipsis
    and venta_mayor is not Ellipsis
    and venta_menor is not Ellipsis
    and rango_ventas is not Ellipsis
), "Complete todas las fórmulas antes de ejecutar el reporte."

print(f"Total: ₡{total_ventas:,.2f}")
print(f"Promedio: ₡{promedio_diario:,.2f}")
print(f"Venta mayor: ₡{venta_mayor:,.2f}")
print(f"Venta menor: ₡{venta_menor:,.2f}")
print(f"Rango: ₡{rango_ventas:,.2f}")
````

## 5. Reto integrador en parejas: Diagnóstico semanal de ventas

Una pequeña empresa registró las ventas de lunes a viernes. La administración
necesita saber si alcanzó su meta semanal, cuánto le faltó o cuánto la superó
y qué información aportan las ventas diarias.

```python
ventas_semana = [145000, 118000, 162000, 137000, 128000]
META_SEMANAL = 700000
```

### Pregunta central

**¿La empresa alcanzó su meta semanal y qué indicadores permiten justificar
la conclusión?**

### Ruta de trabajo

Construyan un programa que:

1. Describa las ventas mediante cantidad, total, promedio, valor mayor,
   valor menor y rango.
2. Exprese qué porcentaje de la meta fue alcanzado.
3. Determine si la meta se cumplió y cuánto faltó o cuánto se superó.
4. Muestre un reporte claro y una conclusión comercial.
5. Conserve los datos originales y utilice nombres descriptivos.

La pareja decide los nombres de las variables, el orden de las instrucciones

### Criterios de evaluación

| Criterio | Peso |
|---|---:|
| Indicadores correctos | 30% |
| Nombres, tipos y clasificación | 20% |
| Decisión y diferencia respecto a la meta | 20% |
| Pruebas del algoritmo | 15% |
| Interpretación de los resultados | 15% |

````python
# RETO INTEGRADOR
ventas_semana = [145000, 118000, 162000, 137000, 128000]
META_SEMANAL = 700000

# Construyan aquí el diagnóstico semanal.
# La solución debe responder la pregunta central y cumplir los requisitos.
cantidad_dias = len(ventas_semana)
total_ventas = sum(ventas_semana)
promedio_diario = total_ventas / cantidad_dias
venta_mayor = max(ventas_semana)
venta_menor = min(ventas_semana)
rango_ventas = venta_mayor - venta_menor

porcentaje_meta = (total_ventas / META_SEMANAL) * 100

diferencia_meta = total_ventas - META_SEMANAL

meta_superada = total_ventas > META_SEMANAL
meta_igualada = total_ventas == META_SEMANAL
meta_cumplida = total_ventas >= META_SEMANAL

assert (
    cantidad_dias is not Ellipsis
    and total_ventas is not Ellipsis
    and promedio_diario is not Ellipsis
    and venta_mayor is not Ellipsis
    and venta_menor is not Ellipsis
    and rango_ventas is not Ellipsis
    and porcentaje_meta is not Ellipsis
    and diferencia_meta is not Ellipsis
    and meta_cumplida is not Ellipsis
), "Complete todas las fórmulas antes de ejecutar el reporte."

print(f"Cantidad de días: {cantidad_dias}")
print(f"Total de ventas: ₡{total_ventas:,.2f}")
print(f"Promedio diario: ₡{promedio_diario:,.2f}")
print(f"Venta mayor: ₡{venta_mayor:,.2f}")
print(f"Venta menor: ₡{venta_menor:,.2f}")
print(f"Rango de ventas: ₡{rango_ventas:,.2f}")
print(f"Porcentaje de la meta alcanzado: {porcentaje_meta:.2f}%")
print(f"Diferencia respecto a la meta: ₡{diferencia_meta:,.2f}")

if meta_superada:
    print(f"Conclusión: la empresa SUPERÓ la meta semanal por ₡{diferencia_meta:,.2f}")
elif meta_igualada:
    print("Conclusión: la empresa IGUALÓ exactamente la meta semanal")
else:
    print(f"Conclusión: la empresa NO ALCANZÓ la meta semanal. Faltaron ₡{abs(diferencia_meta):,.2f}")
````
