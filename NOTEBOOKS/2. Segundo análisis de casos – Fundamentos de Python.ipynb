# Problema #1: Calculadora de propina inteligente

## Problema

Una persona termina de comer en un restaurante y necesita calcular cuánto debe dejar de propina y cuál es el total final a pagar, dependiendo del porcentaje de propina que decida dar.

## Entrada

- Monto total de la cuenta (float)
- Porcentaje de propina deseado (float)

## Proceso

- propina = cuenta * (porcentaje / 100)
- total_final = cuenta + propina

## Salida esperada

- Mensaje indicando el monto de la propina y el total final a pagar, usando f-strings.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | cuenta=10000, propina=10  | Propina: 1000.0, Total: 11000.0 | Propina: 1000.0, Total: 11000.0 |
| Caso límite | cuenta=10000, propina=0 | Propina: 0.0, Total: 10000.0 | Propina: 0.0, Total: 10000.0  |
| Caso inválido | cuenta="diez mil"  | Mensaje de error controlado | Error: debe ingresar un número válido |


```python
# Escriba la solución aquí
try:
    cuenta = float(input("Ingrese el monto total de la cuenta: "))
    porcentaje = float(input("Ingrese el porcentaje de propina (ej: 10): "))

    propina = cuenta * (porcentaje / 100)
    total_final = cuenta + propina

    print(f"Propina a dejar: {propina}")
    print(f"Total final a pagar: {total_final}")

except ValueError:
    print("Error: debe ingresar un número válido")
```

# Problema #2: Detector de número par/impar

## Problema

Se necesita un programa que reciba un número entero y determine si es par o impar, usando el operador módulo.

## Entrada

- Un número entero (int)

## Proceso

- Se calcula numero % 2
- Si el residuo es 0, el número es par; de lo contrario, es impar.

## Salida esperada

- Mensaje indicando si el número ingresado es par o impar.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | numero=7 | 7 es impar | 7 es impar |
| Caso límite | numero=0 | 0 es par | 0 es par |
| Caso inválido | numero="siete" | Mensaje de error controlado | Error: debe ingresar un número entero válido |


```python
try:
    numero = int(input("Ingrese un número entero: "))

    if numero % 2 == 0:
        print(f"{numero} es par")
    else:
        print(f"{numero} es impar")

except ValueError:
    print("Error: debe ingresar un número entero válido")
```

# Problema #3: Clasificador de edad

Se necesita un programa que solicite la edad de una persona y la clasifique en una de tres categorías: menor de edad, adulto o adulto mayor.

## Entrada

- Edad de la persona (int)

## Proceso

- Si edad < 18: menor de edad
- Si edad >= 18 y edad < 65: adulto
- Si edad >= 65: adulto mayor

## Salida esperada

- Mensaje indicando la categoría correspondiente a la edad ingresada.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | edad=30 | Es un adulto | Es un adulto |
| Caso límite | edad=65 | Es un adulto mayor | Es un adulto mayor |
| Caso inválido | edad=-5 | Mensaje de edad inválida | Error: la edad no puede ser negativa |


```python

try:
    edad = int(input("Ingrese su edad: "))

    if edad < 0:
        print("Error: la edad no puede ser negativa")
    elif edad < 18:
        print("Es un menor de edad")
    elif edad < 65:
        print("Es un adulto")
    else:
        print("Es un adulto mayor")

except ValueError:
    print("Error: debe ingresar un número entero válido")
```
